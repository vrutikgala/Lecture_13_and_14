```python
# setup

import pandas as pd
```

```python
#
sales_2017_df=pd.read_csv("Australian_Grocery_Store_Data_2017.csv")
sales_2017_df.sort_values("receipt_id")
```

```python
sales_2018_df=pd.read_csv("Australian_Grocery_Store_Data_2018.csv")
sales_2018_df.sort_values("receipt_id")
```

```python
sales_df=pd.concat([sales_2017_df,sales_2018_df],ignore_index= False)
sales_df.sort_values("receipt_id")
```

```python
item_categories_df=pd.read_csv("Australian_Grocery_Store_Data_Only_Items_Categories.csv")
item_categories_df
```

```python
sales_df=pd.merge(sales_df,item_categories_df, how="left" , on="item_name")
sales_df.head(10)
```

```python
# Making my columns into title case
sales_df["item_name"]=sales_df["item_name"].str.title()
sales_df["payment_type"]=sales_df["payment_type"].str.title()
sales_df["main_category"]=sales_df["main_category"].str.title()
sales_df["sub_category"]=sales_df["sub_category"].str.title()
sales_df.sort_values("receipt_id").head(10)




```

## Make following columns 

```python
sales_df["total_buying_price"]=sales_df["quantity"] * sales_df["unit_buying_price"]
sales_df["total_selling_price"]=sales_df["quantity"] * sales_df["unit_selling_price"]
sales_df["total_margin"]=sales_df["quantity"] * sales_df["unit_price_margin"]
sales_df.sort_values("receipt_id").head(10)
```

```python
# Adding quantity_type column

sales_df.loc[sales_df["quantity"]<=1 , "quantity_type"]= "Single"
sales_df.loc[(sales_df["quantity"]>1)&(sales_df["quantity"]<10) , "quantity_type"]= "Regular"
sales_df.loc[sales_df["quantity"]>=10 , "quantity_type"]= "Bulk"
sales_df.sample(10)
```

```python
sales_df.dtypes
```

```python
sales_df["date"]=pd.to_datetime(sales_df["date"])
sales_df.dtypes
```

```python
# Adding time columns

sales_df["year"]=sales_df["date"].dt.year
sales_df["quarter"]=sales_df["date"].dt.quarter
sales_df["month"]=sales_df["date"].dt.month_name()
sales_df["day"]=sales_df["date"].dt.day_name()
sales_df
```

# Item Analysis

```python
item_df=sales_df.groupby(["item_name","main_category","sub_category"]).agg(unit_buying_price=("unit_buying_price","mean"),
                                                                           unit_selling_price=("unit_selling_price","mean"),
                                                                           unit_price_margin=("unit_price_margin","mean"),
                                                                           quantity=("quantity", "sum"),
                                                                           revenue=("total_selling_price","sum"),
                                                                           profit=("unit_price_margin","sum"))
item_df
```

# What are my top 50 items by revenue?

```python
top_50_item_revenue = item_df.sort_values("revenue",ascending= False).head(50)
top_50_item_revenue
```

# What are my top 50 items by quantity?

```python
top_50_item_quantity=item_df.sort_values("quantity",ascending=False).head(50)
top_50_item_quantity
```

# What are my top 50 items by profit?

```python
top_50_item_profit=item_df.sort_values("profit",ascending= False).head(50)
top_50_item_profit
```

```python
# Reseting the index of item_df
item_df=item_df.reset_index()
item_df
```

# Pantry staples section investigation:
- What is the revenue generated by pantry staples section
- What is the total buying price in the pantry staples section?
- What was the total profit in the Pantry Staples section?

```python
# Make our pantry items_df

pantry_items_df=item_df.loc[item_df["main_category"]=="Pantry Staples"]
pantry_items_df.head(5)
```

```python
# Calculate the revenue , investment , profit for the pantry staples section?

total_revenue=pantry_items_df.revenue.sum()
total_profit=pantry_items_df.profit.sum()
total_investment=total_revenue-total_profit
is_profitable= total_revenue>total_investment

print("Pantry Staples Analysis:")
print("Investment:",total_investment)
print("Revenue:",total_revenue)
print("Profit:",total_profit)
print("Is profitable:",is_profitable)

```

# Which items in Pantry Staples should we stop selling?
- If an item has sold less than 50 units in last 2 years,potentially we may want to consider not selling

```python
items_not_to_keep_df=pantry_items_df.loc[pantry_items_df["quantity"]<50]
pd.set_option("display.max_rows", None)
items_not_to_keep_df.sort_values("quantity")
```

# Which items in Pantry Staples are popular by quantity?

```python
most_popular_by_quantity_df=pantry_items_df.sort_values("quantity",ascending=False)
most_popular_by_quantity_df
```

# Which items in Pantry Staples are making us loose money(negative profit)?

```python
loss_making_df=pantry_items_df.loc[pantry_items_df["profit"]<0]
loss_making_df
```

# Which items in Pantry Staples are making us most profit?

```python
profit_making_df=pantry_items_df.loc[pantry_items_df["profit"]>0].sort_values("profit",ascending= False)
profit_making_df
```

  # Sub Catergory Analysis of the pantry staples category

```python
sub_category_df=pantry_items_df.groupby(["sub_category"]).agg(revenue=("revenue","sum"),
                                                              quantity=("quantity","sum"),
                                                              profit=("profit","sum"))
sub_category_df
```

```python
items_not_to_keep_df.to_csv("Pantry Items Not to Keep.csv")
most_popular_by_quantity_df.to_csv("Most Popular Pantry Items by Quantity.csv")
loss_making_df.to_csv("Pantry Items making Loss.csv")
profit_making_df.to_csv("Pantry Items making Profits.csv")
sub_category_df.to_csv("Sub Category items.csv")
```

```python
# Making a quantile column
item_df["quantile"]=pd.qcut(item_df["quantity"],4 , labels=False)

# Made a dataframe with our upmost quantile only
upmost_quantile_products=item_df.loc[item_df["quantile"]==3]

# Increasing the margins and analysing them
upmost_quantile_products["new_margin"]=upmost_quantile_products["unit_price_margin"]* 1.3
upmost_quantile_products["margin_difference"]=upmost_quantile_products["new_margin"]-upmost_quantile_products["unit_price_margin"]
upmost_quantile_products.sort_values("margin_difference",ascending=False)
```
